```python
import numpy as np
import numpy.ma as ma
from scipy.ndimage.filters import uniform_filter
import matplotlib.pyplot as plt
from utils import display

# Assuming 'image' is your input image
# Let's create a dummy image for the demonstration
image = np.random.rand(100, 100, 3)

# Assuming 'mask' is your binary mask where the mask area is False and the rest is True
# Let's create a dummy mask for the demonstration
mask = np.ones((100, 100), dtype=bool)
mask[40:60, 40:60] = False

# Create a masked array
masked_image = ma.masked_array(image, mask=np.stack([mask]*3, axis=-1))

# Apply the uniform filter (equivalent to blurring)
blurred_masked_image = uniform_filter(masked_image, size=5)

# The blurred_masked_image is also a masked array
# You can convert it back to a normal numpy array if needed:
blurred_image = ma.filled(blurred_masked_image, 0)

display(image, masked_image, blurred_masked_image, blurred_image)
```
